```python
V1,V2,V3,V4,V5,V7,V8 = sym.symbols('V_1,V_2,V_3,V_4,V_5,V_7,V_8')
R1,R2,R3,R4,R5,R7,R8 = sym.symbols('R_1,R_2,R_3,R_4,R_5,R_7,R_8')
Vi,I1,I2,I3,I4 = sym.symbols('V_i,I_1,I_2,I_3,I_4')

e1 = -Vi + V1 + V4
e2 = -V4 + V3 + V5 + V7
e3 = -V7 + V8
e4 = -V3 + V2
e5 = V1 - I1*R1
e6 = V2 - I2*R2
e7 = V3 - (-I2+I3)*R3
e8 = V4 - (-I3+I1)*R4
e9 = V5 - I3*R5
e10 = V7 - (-I4+I3)*R7
e11 = V8 - I4*R8
```

```python
s = sym.solve([e1,e2,e3,e4,e5,e6,e7,e8,e9,e10,e11],
              [V1,V2,V3,V4,V5,V7,V8,I1,I2,I3,I4])
s
```

```python
s[I1].subs({R1:1000,
            R2:1000,
            R3:1000,
            R5:1000,
            R7:1000,
            R4:1000,
            R8:1000}).simplify()
```
